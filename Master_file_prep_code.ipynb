```python
# below is the data that contains till 18th of feb total downloads and views year wise separately
data_2022_2025_year_wise=pd.read_csv("data_contains_all_info.csv")
data_2022_2025_year_wise
```

```python
data_2022_2025_year_wise[data_2022_2025_year_wise["List"]=="S100"]
```

```python
import pandas as pd


# Aggregate data to ensure no duplicate entries for each (List, Year) combination
df = data_2022_2025_year_wise.groupby(['List', 'Year'], as_index=False).agg({'Yearly_Views': 'sum', 'Yearly_Downloads': 'sum'})

# Pivot the data to get yearly views and downloads in separate columns
df_pivot = df.pivot(index='List', columns='Year', values=['Yearly_Views', 'Yearly_Downloads'])

# Flatten the multi-index columns
df_pivot.columns = [f"{col[1]}_{col[0].split('_')[1]}" for col in df_pivot.columns]

# Reset the index
df_pivot.reset_index(inplace=True)

# Sort columns to maintain the desired format
columns_order = ["List"] + [f"{year}_Views" for year in range(2025, 2021, -1)] + [f"{year}_Downloads" for year in range(2025, 2021, -1)]
df_pivot = df_pivot.reindex(columns=columns_order, fill_value=0)

df_pivot
```

```python
# Function to extract the numeric part of the string
def extract_number(s):
    return int(s[1:])  # Assuming the format is always 'S' followed by a number

# Apply the function to the column and sort
df_pivot['sort_key'] = df_pivot['List'].apply(extract_number)
df_pivot= df_pivot.sort_values(by='sort_key')
```
